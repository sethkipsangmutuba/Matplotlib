# Density and Contour Plots

### Visualizing 3D Data in 2D

To visualize 3D data in 2D, use `plt.contour`, `plt.contourf`, or `plt.imshow`â€”for contour lines, filled regions, or image-like plots, respectively.


```python
# Set up inline plotting and import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Apply a clean plot style
plt.style.use('seaborn-v0_8-white')
```

## Visualizing a Three-Dimensional Function

Our first example demonstrates a contour plot using a function $z = f(x, y)$, using the following particular choice for $f$ .

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

### Creating Contour Plots

Use `plt.contour(x, y, z)` for contour plots, where `x`, `y` define grid positions and `z` defines levels. Generate grids easily using `np.meshgrid`.


```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

Now let's look at this with a standard line-only contour plot:

```python
plt.contour(X, Y, Z, colors='black');
```

### Contour Line Styles and Colors

Negative contours use dashed lines; positive use solid. Add `cmap` for color, and increase detail with more contour levels using `np.linspace()`.


```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

### Using Colormaps and Filled Contours

The `RdGy` colormap is ideal for data diverging around zero. Use `plt.contourf` for smoother fills and `plt.colorbar()` to display color scale alongside the plot.


```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

### Smoother Color Mapping with `imshow`

Colorbars help interpret peaks and valleys. To avoid blocky contour fills, use `plt.imshow` with the `interpolation` parameter for a smooth, efficient 2D visualization of the data.


```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy',
           interpolation='gaussian', aspect='equal')
plt.colorbar();
```

### `plt.imshow` Caveats

`imshow` needs manual `extent` for axis scaling, uses top-left origin by default (unlike contour plots), and auto-adjusts aspect ratioâ€”override with `origin` and `aspect` arguments.


### Combining Contours and Images

Overlaying `plt.contour` on a semi-transparent `imshow` (`alpha` controls transparency) combines detail and structure. Use `plt.clabel` to label contour lines directly for clarity.


```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

### Powerful 2D Visualizations of 3D Data

Combining `contour`, `contourf`, and `imshow` offers versatile 3D-to-2D plotting. Explore each functionâ€™s docstring for customization and advanced display options.
